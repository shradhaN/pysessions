# Version Control
---

the task of keeping a software system consisting of many versions and configurations well organized.

**Types**

- Centralized *History is maintained in single centralised location*
    - Subversion (svn)

- Distributed *Each user has their own copy of history*
    - Git (git)
    - Mercurial (hg)

**Free Hosting**

- [Github](https://github.com/) *[ popular ]*
    - git
    - free only for open source projects
- [Gitlab](https://about.gitlab.com/)
    - git
    - free private repositories
- [Bitbucket](https://bitbucket.org/)
    - git
    - hg
    - free private repositories
    - limited to 5 users

## Using Git
---

- Before starting commiting any changes, need to add user information to git config

> `$ git config --global user.name "Diwaker Ghimire"`

> `$ git config --global user.email idwaker@gmail.com`

    >> user information which will be used as author of the commit


## Git

- Clone a repository

> `$ git clone https://github.com/idwaker/pysessions.git`


- Working with the git

> `$ cd pysession`

*show changes made on files*

> `$ git status`


- Adding new file to git 

*tracking new files / create new file named CHANGELOG*
    
> `$ git add CHANGELOG`


- Commiting changes

*commit message is short text that we want to identify this part of history with*

> `$ git commit -m "added CHANGELOG"`


- Listing hostory log

> `$ git log`

- Remote repository

> `$ git remote -v`

    >       origin git@github.com:idwaker/pysessions (fetch)
    >       origin git@github.com:idwaker/pysessions (push)
    
*origin is alias to remote repository*

- Pulling from remote 

> `$ git pull origin master`
    
*master is special branch/lineage on which we are working*

- Pushing our changes to remote

> `$ git push origin master`
    


## Other Git Commands

- Adding all the files to commit

> `$ git add .`

- Adding single file called "README"

> `$ git add README`

- Removing Remote

> `$ git remote rm upstream`


## Hg

- Working with the hg

> `$ hg log`


- Pulling from remote 

> `$ hg pull`
    
*default is special branch/lineage on which we are working*

- Pushing our changes to remote

> `$ hg push`
    

## Maintaining forks
---

Repository forked from https://github.com/idwaker/pysessions.git by user 'student'

- Clone from repository

> `$ git clone https://github.com/student/pysessions.git`

> `$ cd pysessions`

> `$ git remote add upstream https://github.com/idwaker/pysessions.git`

    >> This is only required to run single time.


- Checking remotes
    
> `$ git remote -v`

    >     origin https://github.com/student/pysessions.git (fetch)
    >     origin https://github.com/student/pysessions.git (push)  
    >     upstream https://github.com/idwaker/pysessions.git (fetch)
    >     upstream https://github.com/idwaker/pysessions.git (push)


- Updating our fork from upstream
    
> `$ git checkout master`
    
> `$ git fetch upstream`
    
> `$ git merge upstream/master`
    
> `$ git pull origin master`
    
> `$ git push origin master`

    >> This is required to run everytime, you need to update repository
